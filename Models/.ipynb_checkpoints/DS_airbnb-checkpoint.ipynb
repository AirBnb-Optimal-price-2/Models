```python
def date(df, col):
    """
    This is a function for turning a date column that includes month, day, and year
    into 5 separate columns that includes: Pandas date time, year, month, day, season.
    Seasons identified as: {3,4,5:spring, 6,7,8:summer, 9,10,11:autumn, 12,1,2:winter} 
    """
    df = df.copy()
    season = {1: 'winter',
              2: 'winter',
              3: 'spring',
              4: 'spring',
              5: 'spring',
              6: 'summer',
              7: 'summer',
              8:  'summer',
              9: 'autumn',
              10: 'autumn',
              11: 'autumn',
              12: 'winter'}
    df = pd.DataFrame(data=df)
    df['pd_date_time'] = pd.to_datetime(df[col])
    df['day'] = df['pd_date_time'].dt.day
    df['month'] = df['pd_date_time'].dt.month
    df['year'] = df['pd_date_time'].dt.year
    df['season'] = df['month'].map(season)
    return df
```

```python
#Split the data into traing and testing
training, testing = train_test_split(df136, random_state=42)

target = 'price_x'
drop_col = ['price_y', 'pd_date_time', 'id_x', 'listing_id', 'date', 'reviewer_id',
           'id_y','reviewer_name', 'comments', 'has_avilability']
features = ['room_type_x', 'neighbourhood_group', 'minimum_nights_x', 'accommodates',
           'bathrooms', 'property_type', 'maximum_nights', 'season']

X_train = training[features]
# X_train = training.drop(columns=drop_col)
y_train = training[target]

X_test = testing[features]
# X_test = testing.drop(columns=drop_col)
y_test = testing[target]

print(f"Training Shape:{training.shape}\nTesting Shape:{testing.shape}")
print(f"X_train Shape:{X_train.shape}\ny_train Shape:{y_train.shape}")
print(f"X_test Shape:{X_test.shape}\ny_test Shape:{y_test.shape})")
```

```python
print(X_train.dtypes)
print(df136['minimum_nights_y'].value_counts())
df136.head()
```

```python
#First test on training set
pipeline = make_pipeline(
    ce.OrdinalEncoder(), 
    SimpleImputer(strategy='mean'), 
    StandardScaler(), 
    SelectKBest(f_regression, k=7), 
    XGBRegressor(
        n_estimators=500, 
        n_jobs=-1,
        max_depth=10,
        learning_rate=0.2,
        gamma=0.2,
        objective='reg:squarederror'
    )
)

k = 10
scores = cross_val_score(pipeline, X_train, y_train, cv=k, 
                         scoring='neg_mean_squared_error')
print(f'MSE for {k} folds:', -scores)
```

```python
#The mean RMSE of all scores
print(f'MSE for {k} folds:', -scores)
print(f'RMSE {np.sqrt(-scores.mean())}')
```

```python
pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_test)
print('Gradient Boosting R^2', r2_score(y_test, y_pred))
print('Gradient Boosting RMSE', np.sqrt(mean_squared_error(y_test, y_pred)))
```

```python
# IMPORTANCES

encoder_merge = pipeline.named_steps['ordinalencoder']
encoded_merge = encoder_merge.transform(X_train)
print('X_train_merge shape after encoding', encoded_merge.shape)

model_merge = pipeline.named_steps['xgbregressor']
importances_merge = pd.Series(model_merge.feature_importances_, index=encoded_merge.columns[:8])

n = 7
plt.figure(figsize=(10, n/2))
plt.title(f'Top {n} features')
importances_merge.sort_values()[-n:].plot.barh(color='grey');
```
